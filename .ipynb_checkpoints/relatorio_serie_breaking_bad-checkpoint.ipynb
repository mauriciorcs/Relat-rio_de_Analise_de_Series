```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
episodios = (
    pd.read_csv('dados/avaliacao_episodios_series_top.csv')
    .query("nome_serie == 'Breaking Bad'")
)
episodios
```

```python
episodios[episodios.isnull().T.any()]
```

## Popularidade dos episódios

### Qual a distribuição de votos que cada episódio teve?

```python
sns.histplot(data=episodios, x='num_votos', bins=15, kde=True, color='blue')
plt.title('Distribuição da quantidade de votos por episódio')
plt.xlabel('Quantidade de votos')
plt.ylabel('Frequência')
plt.show()
```

```python
media_votos = episodios['num_votos'].mean()
mediana_votos = episodios['num_votos'].median()
desvio_padrao_votos = episodios['num_votos'].std()
minimo_votos = episodios['num_votos'].min()
maximo_votos = episodios['num_votos'].max()
amplitude = maximo_votos - minimo_votos

Q1 = np.percentile(episodios['num_votos'], 25)
Q3 = np.percentile(episodios['num_votos'], 75)

IQR = Q3 - Q1

# Percentil 25 (primeiro quartil)
p_25 = np.percentile(episodios['num_votos'], 25)


# Percentil 50 (mediana)
p_50 = np.percentile(episodios['num_votos'], 50)

# Percentil 75 (terceiro quartil)
p_75 = np.percentile(episodios['num_votos'], 75)

print(f"Média de votos: {media_votos:.2f}")
print(f"Mediana de votos: {mediana_votos}")
print(f"Desvio padrão de votos: {desvio_padrao_votos:.2f}")
print(f"Mínimo de votos: {minimo_votos}")
print(f"Máximo de votos: {maximo_votos}\n")
print(f"Amplitude dos votos: {amplitude}")
print(f"Intervalo interquartil (IQR): {IQR}\n")

print("Primeiro quartil (25º percentil):", p_25)
print("Mediana (50º percentil):", p_50)
print("Terceiro quartil (75º percentil):", p_75)
```

Tendo em vista que o Intervalo interquartil (IQR): 9004.5, isso significa que a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1) dos dados é de 9004.5 unidades na escala de dados.

Um IQR de 9004.5 indicaria que a dispersão dos meus dados entre o primeiro quartil (25% dos dados estão abaixo de 24421.25) e o terceiro quartil (75% dos dados estão abaixo de 33425.75) é de 9004.5 votos.

### Qual a temporada mais popular?

Assim calculei a média dos votos por episódio para cada temporada, levando em consideração a diferença no número de episódios entre as temporadas.

Essa abordagem considera a consistência da popularidade em cada temporada, ajustando-a para o número de episódios em cada uma. Isso evita a inconsistência que poderia surgir ao simplesmente somar o número total de votos por temporada, dado o número variável de episódios em cada uma.

```python
popularidade_temporada = episodios.groupby('temporada')['num_votos'].mean()
popularidade_temporada
```

```python
temporada_mais_popular = popularidade_temporada.idxmax()

popularidade_media_mais_popular = popularidade_temporada.max()

print(f"A temporada mais popular é a {temporada_mais_popular} com uma popularidade média de {popularidade_media_mais_popular:.2f} votos por episódio.")
```

### Qual a temporada menos popular?

```python
temporada_menos_popular = popularidade_temporada.idxmin()

popularidade_media_menos_popular = popularidade_temporada.min()

print(f"A temporada menos popular é a {temporada_menos_popular} com uma popularidade média de {popularidade_media_menos_popular:.2f}'votos por episódio.")
```

```python
media_avaliacao = episodios['avaliacao_media'].mean()
mediana_avaliacao = episodios['avaliacao_media'].median()
desvio_padrao_avaliacao = episodios['avaliacao_media'].std()
minimo_avaliacao = episodios['avaliacao_media'].min()
maximo_avaliacao = episodios['avaliacao_media'].max()

amplitude_av = maximo_avaliacao - minimo_avaliacao

Q_1 = np.percentile(episodios['avaliacao_media'], 25)
Q_3 = np.percentile(episodios['avaliacao_media'], 75)

IQR_AV = Q_3 - Q_1

# Percentil 25 (primeiro quartil)
Q_25 = np.percentile(episodios['avaliacao_media'], 25)


# Percentil 50 (mediana)
Q_50 = np.percentile(episodios['avaliacao_media'], 50)

# Percentil 75 (terceiro quartil)
Q_75 = np.percentile(episodios['avaliacao_media'], 75)


print(f"Média de Avaliação: {media_avaliacao:.2f}")
print(f"Mediana de Avaliação: {mediana_avaliacao}")
print(f"Desvio Padrão de Avaliação: {desvio_padrao_avaliacao:.2f}")
print(f"Mínimo de Avaliação: {minimo_avaliacao}")
print(f"Máximo de Avaliação: {maximo_avaliacao}\n")

print(f"Amplitude dos votos: {amplitude_av:.2f}")
print(f"Intervalo interquartil (IQR): {IQR_AV}\n")

print(f"Primeiro quartil (25º percentil):{Q_25:.1f}")
print(f"Mediana (50º percentil): {Q_50:.1f}")
print(f"Terceiro quartil (75º percentil): {Q_75:.1f}")
```

O boxplot permite comparar visualmente a distribuição das avaliações médias entre as temporadas. O que da identificar se há alguma diferença significativa nas avaliações médias entre as temporadas e determinar se as primeiras temporadas são melhor avaliadas do que as últimas.

## Padrões temporais

Explore as avaliações de temporadas e episódios ao longo do tempo.

### As primeiras temporadas são melhor avaliadas do que as últimas?

O gráfico de linhas servirá para identificar se há alguma tendência de melhora ou piora nas avaliações à medida que as temporadas avançam. Isso ajudará a determinar se as primeiras temporadas são de fato melhor avaliadas do que as últimas.

```python
avaliacao_temporada = episodios.groupby('temporada')['avaliacao_media'].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(data=avaliacao_temporada, marker='o', color='blue')
plt.title('Avaliação Média por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Avaliação Média')
plt.grid(True)
plt.show()
```

```python
primeiras_temporadas = episodios[episodios['temporada'] <= episodios['temporada'].median()]['avaliacao_media']
ultimas_temporadas = episodios[episodios['temporada'] > episodios['temporada'].median()]['avaliacao_media']
```

```python
primeiras_temporadas.describe()
```

```python
ultimas_temporadas.describe()
```

Como base nisso podemos observar que as ultimas temporadas são mais bem avaliadas que as primeiras

### Os episódios no fim das temporadas são melhor avaliados do que os do início?

A média é calculada entre todos os episódios iniciais e finais das temporadas disponíveis. 

Com isso pode-se observar que há uma certa relação em que os episodios finais de cada temporada se mostram mais bem avaliados que os do inicio de cada temporada.

```python
media_inicio_temporada_ep_um = episodios[episodios['num_episodio'] == 1].groupby('temporada')['avaliacao_media'].mean()
media_fim_temporada_ep_max = episodios[episodios['num_episodio'] == episodios.groupby('temporada')['num_episodio'].transform(max)].groupby('temporada')['avaliacao_media'].mean()
```

Neste caso selecionei apenas o primeiro e o ultimo episódio de cada tempora para comparar, observa-se que em media os episódios dos fins de temporada são mais bom avaliados com exceção da primeira temporada.

```python
plt.figure(figsize=(10, 6))
plt.plot(media_inicio_temporada_ep_um.index, media_inicio_temporada_ep_um.values, label='Média no início da temporada', marker='o')
plt.plot(media_fim_temporada_ep_max.index, media_fim_temporada_ep_max.values, label='Média no fim da temporada', marker='o')
plt.xlabel('Temporada')
plt.ylabel('Avaliação Média')
plt.title('Avaliação Média no Início e no Fim das Temporadas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
episodios_ordenados = episodios.sort_values(by=['temporada', 'num_episodio'])
ultimos_episodios_por_temporada = episodios_ordenados.groupby('temporada').head(3)
media_inicio_temporada = ultimos_episodios_por_temporada.groupby('temporada')['avaliacao_media'].mean()
print(media_inicio_temporada)
```

```python
episodios_ordenados = episodios.sort_values(by=['temporada', 'num_episodio'])
ultimos_episodios_por_temporada = episodios_ordenados.groupby('temporada').tail(3)
media_fim_temporada = ultimos_episodios_por_temporada.groupby('temporada')['avaliacao_media'].mean()
print(media_fim_temporada)
```

Neste caso selecionei os três primeiros e os três ultimos episódios de cada tempora para comparar, observa-se que em media os episódios dos fins de temporada são mais bom avaliados.

```python
plt.figure(figsize=(10, 6))
plt.plot(media_inicio_temporada.index, media_inicio_temporada.values, label='Média no início da temporada', marker='o')
plt.plot(media_fim_temporada.index, media_fim_temporada.values, label='Média no fim da temporada', marker='o')
plt.xlabel('Temporada')
plt.ylabel('Avaliação Média')
plt.title('Avaliação Média no Início e no Fim das Temporadas')
plt.legend()
plt.grid(True)
plt.show()
```

### Qual é a temporada mais regular? (com menor variação de avaliações)

```python
desvio_padrao_temporadas = episodios.groupby('temporada')['avaliacao_media'].std()
temporada_mais_regular = desvio_padrao_temporadas.idxmin()
print("Desvio padrão das avaliações por temporada:")
print(desvio_padrao_temporadas)
print("\nTemporada mais regular:", temporada_mais_regular)
```

### Qual é a temporada mais instável? (com maior variação de avaliações)

```python
temporada_mais_instavel = desvio_padrao_temporadas.idxmax()

print("Desvio padrão das avaliações por temporada:")
print(desvio_padrao_temporadas)

print("\nTemporada mais instável:", temporada_mais_instavel)
```
